```python
from langchain_huggingface import HuggingFaceEmbeddings
embeddings = HuggingFaceEmbeddings(model_name="all-MiniLM-L6-v2")
```

```python
embeddings.embed_query("Hello AI")
```

```python
# docs -> embed, query -> embed, compare both embeddings using cosine_similarity

# docs 
documents = [
    "What is the capital of USA?",
    "What is the Capital of India?",
    "Who is the prime minister of India?"
]

# Embed
doc_embed = embeddings.embed_documents(documents)
doc_embed
```

```python
# query -> embed

#query
query = "Narendra Modi is the prime minister of india?"

#embed
query_embed = embeddings.embed_query(query)
query_embed
```

| Metric            | Similarity Score Range | Behavior                              |
| ----------------- | ---------------------- | ------------------------------------- |
| Cosine Similarity | \[-1, 1]               | Focuses on angle only |
| L2 Distance       | \[0, )                | Focuses on **magnitude + direction**  |


```python
# Use cosine_similarty and finds the related document
from sklearn.metrics.pairwise import cosine_similarity

cosine_similarity([query_embed],doc_embed)
```

```python
# Lets use another similarity search
from sklearn.metrics.pairwise import euclidean_distances
euclidean_distances([query_embed],doc_embed)
```

### FAISS

```python
import faiss
from langchain_community.vectorstores import FAISS
from langchain_community.docstore.in_memory import InMemoryDocstore
```

##### new Faiss index designed for 384-dimensional vectors. When I search it later, calculate the Euclidean (L2) distance to find the closest matches, and store the vectors in a flat (uncompressed) list for perfect accuracy.

`384 means: The dimension (or length) of the vector embeddings you are storing.`

```python
index = faiss.IndexFlatL2(384)
index
```

```python
vector_store = FAISS(embedding_function = embeddings,
                     index = index,
                     docstore=InMemoryDocstore(),
                     index_to_docstore_id={},
)
vector_store.add_texts(["AI is future","AI is powerful","Dogs are cute"])
```

```python
vector_store.index_to_docstore_id
```

```python
faiss_index_id = 1
docstore_id = vector_store.index_to_docstore_id[faiss_index_id]
```

```python
doc_text = vector_store.docstore.search(docstore_id).page_content
doc_text
```

```python
results = vector_store.similarity_search("Tell me about AI", k=3)
results
```

| Feature               | `Flat`                | `IVF` (Inverted File Index)        | `HNSW` (Graph-based Index)          |
| --------------------- | --------------------- | ---------------------------------- | ----------------------------------- |
| Type of Search     | Exact                 | Approximate (cluster-based)        | Approximate (graph-based traversal) |
| Speed               | Slow (linear scan)    | Fast (search only in top clusters) | Very Fast (graph walk)              |


| Dataset Size              | Recommended Index                 |
| ------------------------- | --------------------------------- |
| UPTO 1L                     | `IndexFlatL2` or `IndexFlatIP`    |
| UPTO 1M                  | `IndexIVFFlat` or `IndexHNSWFlat` |
| > 1M                      | `IndexIVFPQ` or `IndexHNSWFlat`   |
